### Binominal distribution

**Bernoulli trial** (or binominal trial) is a random experiemnt with exaclty two posible outcomes.

Binominal distribution can be describes of the discrete probability of how manu successes $k$ you get when you conduct $n$ independent bernoulli trials, each with a provaviltity $p$ to result in 1 or $q=1-p$ to resust in 0 with PMF

If $x$ is binomially distribution with parameters $p \in[0,1]$ and $j \in N$
$$
\operatorname{Pr}[x=i]=\left(\begin{array}{l}
n \\
k
\end{array}\right) p^{l}(1-p)^{n-k}
$$
where

factor is the binominal coeefficient, which is how many number of different ways k successes can come to be out of n trials.

$$
\left(\begin{array}{l}
n \\
k
\end{array}\right)=\frac{n !}{k !(n-k) !}
$$

It holds $$\mathbb{E} [x] = p n$$

#### Bernoulli Distribution

The Bernoulli distribution is a spesial case of the Binominal distribution, but for a single bernoulli trial only. It is the discrete probability distribution of a trial with 2 possible outcomes, either 1 or 0, with respective probability p and q = 1 - p. 

If $x$ is Bernoulli distributed with probability, $p \in[0,1]$, it takes the values
$$
x=\left\{\begin{array}{ll}
1 & \text { with probability } p, \\
0 & \text { with probability } 1-p .
\end{array}\right.
$$
where k is the possible outcome, and p and q are the respective possibilities.

**Where it uses** Bernoulli distribution is used to model the objects distribution given that the object is either detected (k=1) or missdetected

```python
from scipy.stats import bernoulli
bd = bernoulli(0.7) # Instance of Bernoulli distribution with parameter p = 0.7
X = [0, 1] # Outcome of experiment can take value as 0, 1
plt.figure(figsize=(5,5))
plt.xlim(-1, 2)
plt.bar(X, bd.pmf(X), color='orange', width=0.1)
plt.title('PMF Bernoulli (p=0.7)', fontsize='12'); plt.xlabel('Values of Random Variable X (0, 1)', fontsize='12'); plt.ylabel('Probability', fontsize='15'); plt.show()
```

### Different types of target tracking

- **Point object tracking** - at most a single measurement per object per time step.
- **Extended object tracking** - involves multiple measurements per objecet. Posible to also estimate the shape and size of objcets.
- **Group object tracking** - severla objects treated as a single entity - a group
- **Tracking with multipath propagation** - more than one detection per objects per time stamp Caused by multipath phenomenon. Examples: OTHR (over the horizon radar)
- **Tracking with unresolved measurements** - involves multiple object cause one measurement. Work with automotive radar, e.g. two vehicles travellin near each other with the same speed can get "merged" measurement (as a single target)

# Motion model

The object state is a Markov chain that evolves
as
$$
p\left(x_{k} \mid x_{k-1}\right)=\pi_{k}\left(x_{k} \mid x_{k-1}\right)
$$

For instance, we often assume that
$$
x_{k}=f_{k-1}\left(x_{k-1}\right)+q_{k-1}, \quad q_{k-1} \sim N \left(0, Q_{k-1}\right)
$$
such that
$$
\pi_{k}\left(x_{k} \mid x_{k-1}\right)=N\left(x_{k} ; f_{k-1}\left(x_{k-1}\right), Q_{k-1}\right)
$$

# Measurement model

The object is detected with probability $P^{ D }\left(x_{k}\right)$, and then generates a measurement from
$$
p\left(o_{k} \mid x_{k}\right)=g_{k}\left(o_{k} \mid x_{k}\right)
$$


For instance, we often assume that
$$
o_{k}=h_{k}\left(x_{k}\right)+v_{k}, \quad v_{k} \sim N \left(0, R_{k}\right)
$$
such that
$$
g_{k}\left(o_{k} \mid x_{k}\right)=N\left(o_{k} ; h_{k}\left(x_{k}\right), R_{k}\right)
$$

**Note: object is not always detected.**

We use a matrix (or a sequence) to represent the object detections,
$$
O_{k}=\left\{\begin{array}{ll}
{[]} & \text { if object is undetected, } \\
o_{k} & \text { if object is detected. }
\end{array}\right.
$$
We use $\left|O_{k}\right|$ to denote the number of column vectors in $O_{k}$
Given $x_{k},\left|O_{k}\right|$ is Bernoulli distributed:
$$
\left|O_{k}\right|=\left\{\begin{array}{ll}
1 & \text { with probability } P^{D}\left(x_{k}\right), \\
0 & \text { with probability } 1-P^{D}\left(x_{k}\right) .
\end{array}\right.
$$

Using the matrix notation, we get
$$
p\left(O_{k} \mid x_{k}\right)=\left\{\begin{array}{ll}
1-P^{ D }\left(x_{k}\right) & \text { if } O_{k}=[] \\
P^{ D }\left(x_{k}\right) g_{k}\left(o_{k} \mid x_{k}\right) & \text { if } O_{k}=o_{k}
\end{array}\right.
$$
Note: this captures both the probability of detection and, if detected, the distribution of the detection.
Given $x_{k}$, the set of vectors in $O_{k}$ is a Bernoulli random finite set.

PMBM DENSITY
The Poisson Multi-Bernoulli Mixture density is defined as
$$
P M B M _{k \mid k}\left( x _{k}\right)=\sum_{ x _{k}^{\omega} \biguplus x _{k}^{d}= x _{k}} P _{k \mid k}^{u}\left( x _{k}^{u}\right) M B M _{k \mid k}^{d}\left( x _{k}^{d}\right)
$$
PPP density $P _{k \mid k}^{u}(\cdot)$ for undetected objects, typically with mixture intensity,
$$
\lambda_{k \mid k}^{u}\left(x_{k}\right)=\sum_{t=1}^{N_{k}^{u}} w_{k \mid k}^{u, t} p_{k \mid k}^{u, t}\left(x_{k}\right), \quad\left\{\left(w_{k \mid k}^{u, t}, p_{k \mid k}^{u, t}(\cdot)\right)\right\}_{t=1}^{N_{k}^{u}}
$$
Multi-Bernoulli Mixture density $M B M _{k \mid k}^{d}(\cdot)$ for detected objects, with parameters
$$
\left\{\left(\ell_{k \mid k}^{h_{k}},\left\{\left(r_{k \mid k}^{i, h_{k}}, p_{k \mid k}^{i, h_{k}}(\cdot)\right)\right\}_{i=1}^{N_{k}^{h_{k}}}\right)\right\}_{h_{k}=1}^{ H _{k}}
$$
PMBM density parameterised by the mixture intensity parameters, the MB log-weights, and the Bernoulli parameters of the MBs.

3ERNOULLI RECYCLING
Bernoulli recycling: basic idea
Instead of pruning Bernoullis with small $r$, approximate them as PPP, and add the intensity to undetected PPP intensity.
- KL-div minimising PPP approximation: intensity $=$ Bernoulli PHD,
$$
\lambda_{k \mid k}^{h, i, REC }\left(x_{k}\right)=r_{k \mid k}^{h, i} p_{k \mid k}^{h, i}\left(x_{k}\right)
$$
- Undetected PPP intensity after recycling is
$$
\lambda_{k \mid k}^{ u , REC }\left(x_{k}\right)=\lambda_{k \mid k}^{ u }\left(x_{k}\right)+\sum_{h} \sum_{i: r_{k \mid k}^{h, i}<\Gamma^{r}} \exp \left(\ell_{k \mid k}^{h}\right) r_{k \mid k}^{h, i} p_{k \mid k}^{h, i}\left(x_{k}\right)
$$
Note that we must take the normalized hypothesis weight $\exp \left(\ell_{k \mid k}^{h}\right)$ into account

WHY RECYCLING?
Following the recycling, we have to reduce the undetected PPP intensity.
Why not just prune right away?
Recycling/pruning lowers the computational complexity, because there are fewer Bernoullis to consider in the data association.
- Pruning means that we lose all information contained in what is pruned.
By recycling, the information is retained approximately as a PPP.
- The Bernoulli recycling threshold can therefore be considerably larger than a Bernoulli pruning threshold.
- Empirical studies show that Bernoulli recycling leads to lower computational cost, without sacrificing tracking performance.

EXPANDING A BERNOULLI TO CERTAIN EXISTENCE
A Bernoulli $(r, p(\cdot))$ with $r \in(0,1)$ represents two possibilities,
- With probability $r$ we have exactly one object, with state pdf $p(\cdot)$.
- With probability $1-r,$ we have exactly zero objects.
We can represent this as a Bernoulli mixture density,
$$
p( x )=r B ^{1}( x )+(1-r) B ^{2}( x )
$$
where $B ^{1}( x )$ and $B ^{2}( x )$ are Bernoulli densities with parameters
$$
\left(r^{1}, p^{1}(\cdot)\right)=(1, p(\cdot)), \quad\left(r^{2}, p^{2}(\cdot)\right)=(0, \text { any pdf })
$$
- Instead of one Bernoulli with uncertain existence $r,$ we have two hypotheses that each have certain existence: either one object $\left(r^{1}=1\right)$ or zero objects $\left(r^{2}=0\right)$.

PROS AND CONS OF USING LABELS TO FORM TRAJECTORIES
Pros
- Simple and intuitive idea
- Unique labels are a way to try to capture the object identity, which is unique
- Requires little modification to add labels to an unlabelled tracking algorithm
Cons
- Typically, the labels are implicit. The true object identity (the explicit label) is not measured by the sensor(s), it is unobservable.
- Missed detections can lead to gaps in the trajectory formed by labelled estimates
Physically unrealistic switching when multi Bernoulli birth is i.i.d., and in challenging scenarios when objects get in close proximity and then separate
- Difficulty fusing labelled estimates from different sensors

- Disappearing objects, two cases:
